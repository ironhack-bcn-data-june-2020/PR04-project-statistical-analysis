```python
#let's now take a look to the data we are interested in. 

#first, let's look at the distribution of the tournaments 
esports_historical.head(3)
```

```python
#To do that, we are interested in knowing which year had more tournaments happening so we are going to create a categorical column with the year

esports_historical['Year'] = esports_historical['Date'].str.extract(r'(\d{4})')
```

```python
esports_historical
```

```python
#let's organize the columns 
esports_historical.head(3)
esports_historical.info() #Year is an object so we need to change that to a float. 
esports_historical.astype({"Year": int})
```

```python
esports_historical.Year.value_counts()
```

```python
fig,ax = plt.subplots(1, 1, figsize=(12, 6))

ax.hist(esports_historical['Year'], alpha=1, histtype='bar', ec='black', bins=22)
ax.set_title('Year Distribution of Esports tournaments', fontsize=12)
ax.set_ylabel('Year Distribution of Esports tournaments', fontsize=12)
```

```python
"""There is a huge peak in 2020, and we can see that although it was a continuos development, 
in the years between 2007 and 2009 to 2010 to 2014 there was a big increase.

We can see that the exponential growth has happened from 2010 in advance, so we are going to take those years into
account. From 2010, to 2020, analysis of the esports industry. """ 
```

```python
esports_lastdecade = esports_historical.loc[esports_historical['Year']>= '2010']
```

```python
esports_lastdecade.groupby(by = ["Game","Year"]).sum().sort_values(['Earnings','Tournaments'], ascending=False).head(20)
```

```python
"""If we group by game and year, we can understand what was the total of tournamentes, players and earnings per game
and per year. If we sort the results by biggest earning, then by number of tournaments, we get that the game
where most money was made was Fortnite in 2019, with 375 tournaments and over 70 million dollars of earnings."""
```

```python
"""Let's take a closer look to see for how many years has been Fortnite up and how does its earning compare to the
rest of the games"""
```

```python
"""We can see that in the esports side, there are 3832 players, 593 tournaments played and almost a billion of earnings
Let's take a look more closely to the tournaments itself."""

fortnite = esports_lastdecade.loc[esports_historical['Game']=='Fortnite']
fortnite.head()
```

```python
#the first tournament was in?

fortnite.sort_values('Year', ascending = True).head(1) 
#they happened in 2018, which means that there has only been e-tournaments for two years
```

```python
fig,ax = plt.subplots(1, 1, figsize=(12, 6))

ax.hist(fortnite['Year'], alpha=1, histtype='bar', ec='black', bins=5)
ax.set_title('Year Distribution of Esports tournaments', fontsize=12)
ax.set_ylabel('Year Distribution of Esports tournaments', fontsize=12)
```

```python
fortnite.head(10)
```

```python
#lets compare earnings per game. what is the average price per tournament?
data = (list(fortnite.loc[fortnite["Year"] == '2018', "Earnings"]),
        list(fortnite.loc[fortnite["Year"] == '2020', "Earnings"]),
        list(fortnite.loc[fortnite["Year"] == '2019', "Earnings"]))
```

```python
fig1, ax1 = plt.subplots()
labels = ['2018', '2019', '2020']

#selecting data and editing boxplot
ax1.boxplot(data,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('Earnings by Year')
ax1.set_ylim(0, 9000000)
```
